```python
import fitz
import pandas as pd
import numpy as np
```

```python
EB_indi_doc=fitz.open('/content/EB_indi.pdf')
```

```python
print(EB_indi_doc.page_count)
```

```python
EB_indi_doc[0].find_tables().tables[0].to_pandas()
```

```python
len(EB_indi_doc[48].find_tables().tables)
```

```python
df=EB_indi_doc[0].find_tables().tables[0].to_pandas()
for i in range(1,EB_indi_doc.page_count):
  df_new=EB_indi_doc[i].find_tables().tables[0].to_pandas()
  df = pd.concat([df, df_new], ignore_index=True)
```

```python
df
```

```python
duplicate_rows = df[df.duplicated()]
```

```python
print(duplicate_rows)
```

```python
df.to_csv('EB_indi_csv.csv', index=False)
```
